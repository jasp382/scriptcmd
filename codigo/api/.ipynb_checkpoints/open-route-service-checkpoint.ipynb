```python
import requests
import folium
import json
import IPython.display
```

```python
def data_from_get(url, getParams=None, auth=None):
    """
    Return json from URL - GEST Request
    """
    
    response = requests.get(
        url=url, params=getParams,
        headers={'content-type' : 'application/json'},
        auth=auth
    )
    
    return json.loads(response.text)

def http_to_json(url, ojson=None):
    """
    Data from API Endpoint to JSON File
    """

    data = data_from_get(url)

    if not ojson:
        return data
    
    else:
        with open(ojson, 'w') as ff:
            json.dump(data, ff)

        return ojson
```

```python
def directions(lat_o, lng_o, lat_d, lng_d, modeTransportation='foot-walking'):
    """
    Get Shortest path between two points using Directions service
    
    profile options:
    * driving-car;
    * driving-hgv;
    * cycling-regular;
    * cycling-road;
    * cycling-safe;
    * cycling-mountain;
    * cycling-tour;
    * cycling-electric;
    * foot-walking;
    * foot-hiking;
    * wheelchair.
    
    preference options:
    * fastest, shortest, recommended
    
    format options: geojson, gpx
    
    DOC: https://openrouteservice.org/documentation/#/authentication/UserSecurity
    """
    
    URL = (
        f"{MAIN_URL}directions/{modeTransportation}?api_key={TOKEN}&"
        f"start={lng_o},{lat_o}&end={lng_d},{lat_d}"
    )
    
    data = http_to_json(URL)
    
    return data
```

```python
result = directions(
    pointA_lat, pointA_lng,
    pointB_lat, pointB_lng,
    modeTransportation='foot-walking'
)
```

```python
# Create map (adjust location/zoom)
m = folium.Map(location=[40, -8], zoom_start=6)

# Add GeoJSON
geo_layer = folium.GeoJson(
    result,
    name="Fastest route"
)
geo_layer.add_to(m)


# --- ZOOM TO LAYER ---
bounds = geo_layer.get_bounds()
m.fit_bounds(bounds)

# Display map
m
```
