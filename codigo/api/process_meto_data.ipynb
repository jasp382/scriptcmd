```python
import os
import json
import pandas as pd
```

```python
json_folder = '/home/luffy/code/scriptcmd/dados/ipma_data/'
```

```python
# List Json files
jsfiles = []
for (d, d_, f) in os.walk(json_folder):
    jsfiles.extend(f)
    break

# Percorrer todos os ficheiros JSON do mês
alldata = []
for jsfile in jsfiles:
    jspath = os.path.join(json_folder, jsfile)
    with open(jspath, "r", encoding="utf-8") as f:
        data = json.load(f)

    # Estrutura: timestamp -> estações -> variáveis
    for timestamp, estacoes in data.items():
        for estacao, valores in (estacoes or {}).items():
            if valores is None:
                continue

            row = {
                'daytime'   : timestamp,
                'stationid' : estacao
            }

            for val in valores:
                row[val] = valores[val]

            alldata.append(row)
```

```python
meteo_df = pd.DataFrame(alldata)
```

```python
# Apresentação dos resultados
print("📊 Estatísticas mensais por estação\n")

for estacao in sorted(meteo_df.stationid.unique()):
    filter_df = meteo_df[meteo_df.stationid == estacao]
    media_temp = filter_df.temperatura.mean()
    print(media_temp)
    
    soma_prec = filter_df.temperatura.sum()

    print(
        f"Estação {estacao} | "
        f"Média temperatura: {media_temp:.2f} °C | "
        f"Precipitação total: {soma_prec:.2f} mm"
    )
```
